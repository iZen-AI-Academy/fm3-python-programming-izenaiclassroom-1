This project requires you to implement a set of functions that test your understanding of the Python programming language. The class Matrix, shown below, implements a 2-dimensional array. It has two instance variables: shape and array. 'shape' is a tuple: shape[0] stores the number of rows, and shape[1] stores the number of columns. 'array' stores the actual data in the matrix. It is implemented as a list of lists. For instance, the list [[1, 2], [3, 4], [5, 6]] is a 2 dimensional matrix: it has 3 rows and 2 columns. The first row is [1, 2], the second row [3, 4], and the third row [5, 6]. The class and a constructor for the class are shown below. 

````python
class Matrix:
    
    def __init__(self, *rows): ## constructor
        self.array = []        ## instance variable 'array' -> initialized to empty set
        
        ### find the number of rows in the Matrix object
        num_rows = len(rows)
        num_cols = 0
        
        ### Iterate over all rows and add it to self.array
        for row in rows:
            self.array.append(row)
        
            ### find the maximum size of each list -> becomes column size
            if (num_cols < len(row)):
                num_cols = len(row)
                
        ### set up the 'shape' instance variable
        self.shape=(num_rows, num_cols)
        
        ### make sure that all rows have the same number of elements. If
        ### not, add zero elements.
        for row in self.array:
            cols = len(row)
            if (cols < num_cols):
                for col in range(cols, num_cols):
                    row.append(0.0)
````

A Matrix object is created by passing a list of rows as an argument to the constructor. The code below shows how Matrix objects can be created, and how the different instance variables can be accessed.

    ```
    m = Matrix([1, 2, 3], [4, 6], [83, 24, 77], [11, 99, 27])
    print(m.shape)
    print(m.array)
    -------------------------------------------------
    Output:
    (4, 3)
    [[1, 2, 3], [4, 6, 0.0], [83, 24, 77], [11, 99, 27]]
    ```
In the above, the constructor is passed 4 arguments: [1, 2, 3], [4, 6], [83, 24, 77], [11, 99, 27]. Each denotes a row of 3 elements.
